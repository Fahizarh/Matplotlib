### Creating Simple Plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'k')
```

### Creating Multiplots on the Same Canvas

There are 2 ways of doing this

1. Functional method

2. Object-oriented Method

### Functional method

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--', lw= 2, marker= '+')

plt.subplot(1,2,2)
plt.plot(y,x,'g', marker = 'o')
```

### Object-Oriented Interface Single Plot

```python
# First create a blank figure(area where you can plot)
fig = plt.figure()

# Then add a set of axes using the .add_axes() method
# Syntax = fig.add_axes([left,bottom,width,height]). The values range from 0 to 1
ax = fig.add_axes([0.1,0.2,0.8,0.9])

# You can then plot the graph on the axes and label 
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

### Creating Multiple Figures using the Object-Oriented Interface

```python
fig = plt.figure()

# You can create different axes depending on the number of plots
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y, label= 'First plot' )
ax2.plot(y,x,'y--', label = 'Second plot')

ax1.set_title('Title')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ax2.set_title('Title')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

fig.legend(loc='lower right')
```

### Creating Subplots

```python
# We can create subplots by using the .subplots() method

fig,axes = plt.subplots(nrows=3, ncols=4)

plt.tight_layout()

axes[0,2]= plt.plot(x,y)
axes[1,1]= plt.plot(y,x,'r*')
```

### Figure size, aspect ratio and DPI(dots-per-inch)

```python
# fig = plt.figure(figsize=(width,height), dpi=value)

fig= plt.figure(figsize=(8,2), dpi=100)
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

### Saving Figures

You can save figures by using the .savefig() method. Figures can be saved in different formats(png, jpg, svg, pdf, etc)

```python
fig.savefig('my_figure.jpg')
```
